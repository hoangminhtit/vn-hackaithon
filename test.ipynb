```python
import pandas as pd
```

```python
pred = pd.read_csv("output/pred.csv")
pred.head()
```

```python
result = pd.read_csv("results/test_1780332041.csv")
result.head()
```

```python
submission = pd.DataFrame({
    "qid": result.columns,
    "answer": result.iloc[0].values,
    "domain": result.iloc[1].values
})

submission.head()
```

```python
merge_ = pd.merge(pred, submission, on="qid", how="left")
merge_.head()
```

```python
a = merge_['answer_x'] == merge_['answer_y']
a.value_counts()
```

```python
diff = merge_[merge_['answer_x'] != merge_['answer_y']]

stats_ = diff[['qid', 'domain']]
```
